### 4) Lexical diversity  short explanation

Explain in your own words what aspect of language lexical diversity describes.

Lexical diversity measures how varied the vocabulary of a text is. A common proxy is the typetoken ratio (TTR) = unique word types / total tokens. Higher values indicate greater vocabulary variety; very long texts tend to have lower TTR because repeated words accumulate.

````python
def lexical_diversity(tokens):
    return len(set(tokens)) / len(tokens)

print(f"TTR (text2): {lexical_diversity(text2):.4f}")
````

### 6) What is a bigram? How does it relate to collocations?

What is a bigram? How does it relate to collocations? Explain in your own words.

A bigram is a sequence of two adjacent tokens (e.g., New York). Collocations are bigrams (or longer n-grams) that occur together unusually often, beyond what chance (independent occurrence) would suggest. In NLTK, text.collocations() reports high-association bigrams using association measures (e.g., PMI). Thus, every collocation is a bigram (or n-gram), but not every bigram is a collocation.

````python
print("Top collocations in text1:")
text1.collocations()
````

### Ex 15  Content fraction
Fraction of tokens that are alphabetic and not stopwords.

````python
def content_fraction(tokens):
    stops = set(stopwords.words("english"))
    words_alpha = [w for w in tokens if w.isalpha()]
    content = [w for w in words_alpha if w.lower() not in stops]
    return len(content) / len(tokens)

for tname, t in [("text1", text1), ("text2", text2), ("text4", text4)]:
    print(f"[Ex15] {tname} content fraction = {content_fraction(t):.3f}")
````

### Ex 17  Hapax fraction
Fraction of tokens that occur only once.

````python
def hapax_fraction(tokens):
    fd = FreqDist(w.lower() for w in tokens if w.isalpha())
    return len(fd.hapaxes()) / len(fd)

for tname, t in [("text1", text1), ("text2", text2), ("text5", text5)]:
    print(f"[Ex17] {tname} hapax fraction = {hapax_fraction(t):.3f}")
````

##  Zipfs Law  BarabsiAlbert Networks

**Zipfs law** describes how words are distributed in natural language:  
a few words (like the, of, and) occur **very often**, while most words are **rare**.  
This pattern follows a *power-law distribution*  the frequency of a word is roughly **inversely proportional** to its rank in frequency.

In simple terms:
- Rank 1 (the most common word) might appear thousands of times.
- Rank 10 appears hundreds of times.
- Rank 1000 appears just a few times.

The **BarabsiAlbert (BA)** model in network science shows a similar pattern  it describes how **networks grow** over time when new nodes prefer to connect to already well-connected nodes (a process called *preferential attachment*).

Mathematically:

- In BA networks, the probability that a node has degree *k* follows  
  **P(k)  k**, where **  3**

- In natural language, the frequency of a word with rank *r* follows  
  **f(r)  r**, where **  1**

Both systems are **scale-free**  meaning the same statistical pattern appears at different scales  and both reflect *rich-get-richer* dynamics:
- Popular words are used more, becoming even more frequent.
- Popular nodes (like big websites) get more links, becoming even more central.

###  Other Examples of Power Laws on the Internet
- The number of links per web page (few pages have millions, most have few)
- Number of followers on social media
- File sizes and web traffic distributions
- Popularity of songs, videos, or hashtags (a few go viral, most stay small)

**In short:**  
Zipfs law in language and the BarabsiAlbert model in networks both show that **growth and popularity follow unequal, but predictable, patterns.**

##  Understanding `ConditionalFreqDist` (in simple terms)

`ConditionalFreqDist` is an NLTK tool used to count how often things occur **under different conditions**.  
Its like a frequency table grouped by categories.

Example idea:
> How often do modal verbs like *can*, *could*, *may*, *might* appear in different genres of the Brown corpus?

```python
from nltk.corpus import brown
from nltk import ConditionalFreqDist

cfd = ConditionalFreqDist(
    (genre, word.lower())
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)

# Show frequencies for a few modals
modals = ["can", "could", "may", "might", "must", "will"]
cfd.tabulate(conditions=["news", "romance"], samples=modals)

Think of it as a dictionary of frequency distributions, where each condition has its own FreqDist inside.
